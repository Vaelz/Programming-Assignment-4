```python
import pandas as pd
```

```python
data = pd.read_csv('board2.csv')
data
```

# Problem 1

```python
data['Average'] = data[['GEAS','Electronics','Math', 'Communication']].mean(axis=1)
```

```python
Instru = data[(data['Track'] == 'Instrumentation') & (data['Hometown'] == 'Luzon') & (data['Electronics'] > 70)]
Instru = Instru[['Name','GEAS','Electronics']]
Instru
```

```python
Mindy = data[(data['Hometown'] == 'Mindanao') & (data['Gender'] == 'Female') & (data['Average'] >= 55)]
Mindy = Mindy[['Name','Track','Electronics','Average']]
Mindy
```

# Problem 2

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20, 4))
```

```python
tracks = data['Track'].unique()
track_averages = [data[data['Track'] == track]['Average'] for track in tracks]

plt.boxplot(track_averages, labels=tracks)
plt.title('Contribution of Track to Average Grade')
plt.xlabel('Track')
plt.ylabel('Average')
plt.show()
```

```python
genders = data['Gender'].unique()
gender_averages = [data[data['Gender'] == gender]['Average'] for gender in genders]

plt.boxplot(gender_averages, labels=genders)
plt.title('Contribution of Gender to Average Grade')
plt.xlabel('Gender')
plt.ylabel('Average')
plt.show()
```

```python
hometowns = data['Hometown'].unique()
hometown_averages = [data[data['Hometown'] == hometown]['Average'] for hometown in hometowns]

plt.boxplot(hometown_averages, labels=hometowns)
plt.title('Contribution of Hometown to Average Grade')
plt.xlabel('Hometown')
plt.ylabel('Average Grade')
plt.show()
```
